```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, np.pi, 100)
```

```python
def fn(q, Q):
    return 1 - (1 - (1 - x / np.pi)**q)**Q
```

```python
plt.plot(np.cos(x), fn(24, 1024), c='r')
plt.plot(np.cos(x), fn(16, 1024), c='b')
plt.plot(np.cos(x), fn(12, 1024), c='g')
```

```python
plt.plot(np.cos(x), fn(14, 16384), c='r')
plt.plot(np.cos(x), fn(18, 16384), c='b')
plt.plot(np.cos(x), fn(28, 16384), c='g')
```

```python
o = 35
for h, g, c in [
    (128, 16, 'r'),
    (64, 4096, 'g'),
    (32, 4096, 'b'),
    (28, 16384, 'y')
]:
    plt.plot(np.cos(x)[:o], fn(h, g)[:o], c=c, label=f'h={h}, g={g}')
plt.legend()
ax = plt.gca()
ax.set_title('Hash Collision Probability')
ax.set_xlabel('True Cosine Similarity')
ax.set_ylabel('Probability')
plt.gcf().set_size_inches(16, 4)
```
